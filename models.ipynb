```python
from pycaret.datasets import get_data
data = get_data('Clean_Data/Clean_data44t_Hatyai')
data = data.drop(columns=['Unnamed: 0','Unnamed: 0.1'])
print(data)
```

```python
from pycaret.regression import *
s = setup(data, target = 'PM25', session_id = 123)
```

```python
best = compare_models()
```

```python
evaluate_model(best)
```

```python
predict_model(best)
```

```python
predict_model(best, data=data)
```

```python
save_model(best, 'Final_Model_PM25')
```
